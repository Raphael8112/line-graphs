```python
import matplotlib.pyplot as plt #plotting
import numpy as np #mathematical operations
```

```python
#create the data
year=[2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
profits=[20,25,30,23,68,89,90,54,34,67,80]
```

```python
from matplotlib.lines import lineStyles
#plot the graph
plt.plot(year,profits,linestyle='dashed',marker='o')
plt.title("Profit and Loss")
plt.xlabel("Years")
plt.ylabel("Profit")
plt.show()
```

```python
#create random values from 1 to 10
x=np.arange(1,11,1)
y1=(2*x)+1
y2=(2*x*x)+2
```

```python
#plot the linecharts
plt.plot(x,y1,'g',linewidth=3,label='y=2x+1')
plt.plot(x,y2,'r',linewidth=3,label='y=2x^2+2')
plt.legend()
plt.show()

```
